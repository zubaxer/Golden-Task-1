```python
# Importing the required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
df = pd.read_csv('portfolio_data.csv')
df.head()
```

```python
df.info()
df.describe()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
print(df.info())
print(df.describe())
```

```python
# Plot the time series for each stock
plt.figure(figsize=(14, 8))
plt.plot(df.index, df['AMZN'], label='Amazon')
plt.plot(df.index, df['DPZ'], label='Domino\'s Pizza')
plt.plot(df.index, df['BTC'], label='Bitcoin')
plt.plot(df.index, df['NFLX'], label='Netflix')
plt.title('Stock Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Adjusted Price')
plt.legend()
plt.show()
```

```python
rolling_window = 30
df_rolling_mean = df.rolling(window=rolling_window).mean()
df_rolling_std = df.rolling(window=rolling_window).std()

plt.figure(figsize=(14, 8))
for column in df.columns:
    plt.plot(df_rolling_mean.index, df_rolling_mean[column], label=f'{column} Rolling Mean')
    plt.plot(df_rolling_std.index, df_rolling_std[column], linestyle='--', label=f'{column} Rolling Std Dev')
plt.title('Rolling Mean and Volatility')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
correlation_matrix = df.corr()
print("Correlation Matrix:")
print(correlation_matrix)
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

for column in df.columns:
    result = seasonal_decompose(df[column], model='multiplicative', period=365)
    result.plot()
    plt.title(f'Seasonal Decomposition of {column}')
    plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA

for column in df.columns:
    model = ARIMA(df[column], order=(5, 1, 0))
    results = model.fit()
    forecast = results.forecast(steps=30)
    plt.figure(figsize=(10, 6))
    plt.plot(df[column], label='Historical Data')
    plt.plot(forecast, label='Forecast')
    plt.title(f'ARIMA Forecast for {column}')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.show()
```
